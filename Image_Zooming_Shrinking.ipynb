```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = cv2.imread('monalisa.jpeg')
```

```python
cv2.imshow("Monalisa",img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
a, b = img.shape[:2]
```

```python
def ImageVerticalFlip(img):
    return plt.imshow(img[::-1,:,::-1])
```

```python
def ImageHorizontalFlip(img):
    return plt.imshow(img[:,::-1,::-1]);
```

```python
# to zoom rgb colors we have to use three loops
def Zoom(img,z):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    zoomedImg = np.zeros((a*z,b*z))
    for i in range(a):
        for j in range(b):
            zoomedImg[(i-1)*z, (j-1)*z] = img[i,j]
    return plt.imshow(zoomedImg)
```

```python
# to shrink rgb colors we have to use three loops
def Shrink(img,z):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ShrinkedImg = np.zeros((a//z,b//z))
    for i in range(a//z):
        for j in range(b//z):
            ShrinkedImg[i,j] = img[(i-1)*z, (j-1)*z]
    return plt.imshow(ShrinkedImg)
```

```python
print("""
1. Vertically flip the image
2. Horizontally flip the image
3. Zooming of image
4. Shrinking of image
5. Dispaly image
""")
choice = int(input('Choose your desired operation:'))

if choice == 1:
    ImageVerticalFlip(img)
elif choice == 2:
    ImageHorizontalFlip(img)
elif choice == 3:
    z = int(input('Zooming factor: '))
    Zoom(img, z)
elif choice == 4:
    z = int(input('Shrinking factor: '))
    Shrink(img, z)
else:
    plt.imshow(img[:,:,::-1])
```
